```python
import csv
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
file1 = open('Raw_Skills_Dataset.csv')
file2 = open('Example_Technical_Skills.csv')
```

```python
rawData = csv.reader(file1)
techSkills = csv.reader(file2)
```

```python
header1 = next(rawData)
header1
```

```python
header2 = next(techSkills)
header2
```

```python
technologySkills = []
for i in techSkills:
  technologySkills.append(i[0].lower())
techSkillsSet = set(technologySkills)
techSkillsSet
```

```python
rawDataSkills = []
for i in rawData:
  rawDataSkills.append(i[0].lower().strip())
```

```python
tempList1 = []
for i in rawDataSkills:
  for j in techSkillsSet:
    if j in i:
      tempList1.append(i)
      break
tempSet1 = set(tempList1)
```

```python
list1 = list(tempSet1)
tempList2 = []
for i in list1:
  if ('.com' and '@') not in i:
    i = i.replace('/','\n')
    i = i.replace(',','\n')
    tempList2.append(i)
tempSet2 = set(tempList2)
tempSet2
```

```python
with open(r'hardSkills.txt', 'w') as fp:
    for i in tempSet2:
        # write each item on a new line
        fp.write("%s\n" % i.strip(' '))
```
